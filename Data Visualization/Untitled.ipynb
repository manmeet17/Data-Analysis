```python
ds=randn(25)
```

```python
sns.rugplot(ds)
plt.ylim(0,1)
```

```python
plt.hist(ds,alpha=0.3)
sns.rugplot(ds)
```

```python
sns.rugplot(ds)
xmin=ds.min()-2
xmax=ds.max()+2
xaxis=np.linspace(xmin,xmax,100)
band=((4*(ds.std()**5))/(3*len(ds)))**0.2
kernel=list()
for i in ds:
    k=stats.norm(i,band).pdf(xaxis)
    kernel.append(k)
    
    k=k/k.max()
    k=k*0.4
    
    plt.plot(xaxis,k,color='grey',alpha=0.5)
    
plt.ylim(0,1)
```

```python
skde=np.sum(kernel,axis=0)
kernel
```

```python
fig=plt.plot(xaxis,skde,color='indianred')
sns.rugplot(ds)
plt.yticks([])
plt.suptitle("Sum of the basis functions")
```

```python
sns.kdeplot(ds)
```

```python
sns.rugplot(ds,color='black')
for bw in np.arange(0.5,2,0.25):
    sns.kdeplot(ds,bw=bw,lw=1.8,label=bw)
```

```python
kernel_options = ["biw", "cos", "epa", "gau", "tri", "triw"]
for kern in kernel_options:
    sns.kdeplot(ds,kernel=kern,label=kern,shade=1)
```
