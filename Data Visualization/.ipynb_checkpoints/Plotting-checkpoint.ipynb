```python
ds=randn(25)
```

```python
sns.rugplot(ds)
plt.ylim(0,1)
```

```python
plt.hist(ds,alpha=0.3)
sns.rugplot(ds)
```

```python
sns.rugplot(ds)
xmin=ds.min()-2
xmax=ds.max()+2
xaxis=np.linspace(xmin,xmax,100)
band=((4*(ds.std()**5))/(3*len(ds)))**0.2
kernel=list()
for i in ds:
    k=stats.norm(i,band).pdf(xaxis)
    kernel.append(k)
    
    k=k/k.max()
    k=k*0.4
    
    plt.plot(xaxis,k,color='grey',alpha=0.5)
    
plt.ylim(0,1)
```

```python
skde=np.sum(kernel,axis=0)
kernel
```

```python
fig=plt.plot(xaxis,skde,color='indianred')
sns.rugplot(ds)
plt.yticks([])
plt.suptitle("Sum of the basis functions")
```

```python
flights=sns.load_dataset("flights")
```

```python
flights
```

```python
flights=flights.pivot("month","year","passengers")
```

```python
flights
```

```python
sns.heatmap(flights)
```

```python
sns.heatmap(flights,annot=True,fmt='d')
```

```python
sns.heatmap(flights,center=flights.loc['August',1960])
```

```python
from pandas import DataFrame
f, (axis1,axis2) = plt.subplots(2,1)
year_fl = flights.sum()
year=Series(year_fl.index.values)
year=DataFrame(year)

fl=Series(year_fl.values)
fl=DataFrame(fl)

yr_fl=pd.concat((year,fl),axis=1)
yr_fl.columns=['Year','Flights']
sns.barplot('Year','Flights',data=yr_fl,ax=axis1)
sns.heatmap(flights,cmap='Blues',ax=axis2,cbar_kws={'orientation':'horizontal'})
```
